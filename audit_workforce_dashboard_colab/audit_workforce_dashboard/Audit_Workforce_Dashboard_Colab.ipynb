## 1. Upload the project ZIP
Upload `audit_workforce_dashboard_colab.zip` or `audit_workforce_dashboard.zip`.


```python
from google.colab import files
uploaded = files.upload()
zip_name = next(iter(uploaded.keys()))
zip_name
```

## 2. Extract the project and find `app.py` automatically


```python
import os, shutil, zipfile

extract_dir = '/content/audit_dashboard_project'
shutil.rmtree(extract_dir, ignore_errors=True)
os.makedirs(extract_dir, exist_ok=True)

with zipfile.ZipFile(zip_name, 'r') as z:
    z.extractall(extract_dir)

project_dir = None
for root, dirs, files_ in os.walk(extract_dir):
    if 'app.py' in files_ and 'requirements.txt' in files_:
        project_dir = root
        break

if project_dir is None:
    raise FileNotFoundError('Could not find app.py and requirements.txt inside the ZIP.')

print('Project folder:', project_dir)
%cd $project_dir
```

## 4. Launch the dashboard
After running this cell, Colab will open the Streamlit dashboard. Use the sidebar to change country and working hours.


```python
import subprocess, time
from google.colab import output

# Stop old Streamlit process if any
!pkill -f streamlit || true

process = subprocess.Popen([
    'streamlit', 'run', 'app.py',
    '--server.port', '8501',
    '--server.headless', 'true',
    '--server.enableCORS', 'false',
    '--server.enableXsrfProtection', 'false'
], stdout=subprocess.PIPE, stderr=subprocess.STDOUT, text=True)

time.sleep(5)
output.serve_kernel_port_as_window(8501)
```

Run the next cell only if Streamlit does not open.


```python
# Print recent Streamlit output if needed
try:
    for _ in range(20):
        line = process.stdout.readline()
        if not line:
            break
        print(line, end='')
except Exception as e:
    print(e)
```
